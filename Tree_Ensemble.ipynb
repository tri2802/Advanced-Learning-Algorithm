**Randomizing the Feature Choice**:  
- At each node, when choosing a feature to use to split, if n features are available, pick a random subset of k < n features and allow the algorithm to only choose from that subset of features
- By convention, choose $k = \sqrt{n}$

#####Part V: When to Use Decision Tree

**Decision Tree and Tree Ensemble**
- Works well on tabular (structured) data
- Not reccomended for unstructured data (image, audio, text)
- Fast
- May be human-interpretable

**Neural Network**
- Works well on all type of data
- May be slower
- Works with transfer learning
- When building a system of multiple models working together, it might be easier to string together multiple neural network
