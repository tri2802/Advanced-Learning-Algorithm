# Part I: Alternative to the Sigmoid Activation

**Sigmoid**

$$g(z) = \frac{1}{1 + e^{-z}}$$

**ReLU**

$$g(z) = max(0, z)$$

**Linear (no activation)**

$$g(z) = z$$

# Part II: Choosing Activation Functions

**Output Layer**
- Choosing the activation function for the output layer would be corresponds to the characteristics of the response training variable
- For instance:  
Binary Classification: sigmoid, since y is 0 or 1  
Regression: linear, y is negative or positive  
Regression: reLU, 0 or +

**Hidden Layer**

- reLU is the most common activation function for the hidden layers
- Reasons:  
reLU is generally faster, no need to compute the exponential 
reLU is flat on one end while sigmoid is flat on both ends, therefore, running gradient descent on reLU is more effective

# Part III: Why Do We Need Activation Functions

**Linear Example**


$$a^{[1]} = w_{1}^{[1]} x + b_{1}^{[1]}$$
$$a^{[2]} = w_{1}^{[2]} a^{[1]} + b_{2}^{[1]}
         = (w_{1}^{[1]} w_{1}^{[2]}) + w_{1}^{[2]} b_{1}^{[1]} + b_{1}^{[2]}$$

$a^{[2]}$ is now equivalent to $wx + b$ by setting w and b accordingly, this is just linear regression

**Example**


- For this example, if all the layers use linear activation function, the result will still be linear, and this is equivalent to using linear regression
- If the output layer were to use sigmoid activation, while the hidden layers are still linear, then this would be equivalent to using logistic regression
- Rule of thumb: Don't use linear regression in hidden layers
