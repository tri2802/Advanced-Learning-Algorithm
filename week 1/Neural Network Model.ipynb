## Part I: Neural Network Layer

**Recall**:
- We have an input layer, denoted as layer 0, contains a vector of 4 values
- We want to create a layer of neural network of 3 neurons, and perform logistic regression on it  

**Then**:  
- The output of layer 1, denoted as a vector of 3 values, becomes the input for layer 2
- The notation: $\vec{a^{[i]}}$ denoted the output vector of the $i^{th}$ layer

- Now, the vector $[0.3, 0.7, 0.2]$ becomes the input for layer two
- Performing logistic regression on this layer gives the value of $\vec{a^{[2]}}$ to be 0.84
- Optional: binary prediction, is $\vec{a^{[2]}} >= 0.5$, then yes or no
