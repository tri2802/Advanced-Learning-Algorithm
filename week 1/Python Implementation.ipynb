# Part I: Forward Propagation in a Single Layer

```python
import numpy as np
```

```python
#Defining the sigmoid function
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
x = np.array([200, 17]) #input vector
```

**First Layer**


```python
#First Param
w1_1 = np.array([1, 2])
b1_1 = np.array([-1])
z1_1 = np.dot(w1_1, x) + b1_1
a1_1 = sigmoid(z1_1)

#Second Param
w1_2 = np.array([-3, 4])
b1_2 = np.array([1])
z1_2 = np.dot(w1_2, x) + b1_2
a1_2 = sigmoid(z1_2)

#Third Param
w1_3 = np.array([5, -6])
b1_3 = np.array([2])
z1_3 = np.dot(w1_3, x) + b1_3
a1_3 = sigmoid(z1_3)

#Stacking the results of first layer
a1 = np.array([a1_1, a1_2, a1_3])
```

**Second Layer**


```python
w2_1 = np.array([-7, 8, 9])
b2_1 = np.array([3])
z2_1 = np.dot(w2_1, a1) + b2_1
a2_1 = a2 = sigmoid(z2_1)
```

## Part II: General Implementation of Forward Propagation

```python
#We stack the weights and bias of the first layer into each column of a matrix
W1 = np.array([
    [1, -3, 5],
    [2, 4, -6]
])

#b is a row vector
b1 = np.array([-1, 1, 2])

#define the input vector
x = a_in = np.array([-2, 4])
```

```python
#Defining the dense function in logistic regression
def dense(a_in, W, b):
    units = W.shape[1] #the number of columns
    a_out = np.zeros(units) #initialize the output of the layer
    
    for j in range(units):
        w = W[:, j] #accessing each column
        z = np.dot(a_in, w) + b[j]
        a_out[j] = sigmoid(z)
    
    return a_out
```

```python
#Defining the method to stack each layer
def sequential(x): #x is an input vector
    a1 = dense(x, W1, b1)
    a2 = dense(a1, W2, b2)
    a3 = dense(a2, W3, b3)
    
    #assuming we only use 3 layers
    f_x = a3
    
    return f_x
```
