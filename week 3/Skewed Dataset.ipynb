# Part I: Error Metrics in Skewed Datasets

**Rare Disease Classification Example**  
Train classifier$f_{\vec{w}, b}(\vec{x})$: y = 1 if disease present, 0 otherwise  
Find that we got 1% error on test set (99% accuracy)  
Only 0.5% patient have the disease  
If our algorithm is: print("y = 0"), we would still get 0.5% error; however, this may not be the optimal model for prediction

**Precision/Recall**


- Precision (of all the people we predicted 1, what fraction is actually correct):  
$$\frac{True positive}{predicted positive} = \frac{True positive}{True positive and False positive}$$   
- Recall (of all people actually is 1, what fraction we correctly predicted):  
$$\frac{True positive}{actual positive} = \frac{True positive}{True positive and False negative}$$
