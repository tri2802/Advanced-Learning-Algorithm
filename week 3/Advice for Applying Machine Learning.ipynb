# Part I: Evaluating a Model

**Evaluating Your Model**


Model fits training data well but will fail to generalize to new examples not in the training set


- Split the dataset into training and testing set
- Formally, training set will take 70% of the data, and 30% remains for testing

**Procedure for Linear Regression** (with squared error cost)

When computing the cost for the training or testing set, omit the regularization term


- In this example of overfitting, $J_{train}(\vec{w}, b)$ will be low  
- $J_{test}(\vec{w}, b)$ will correspondingly be high

**Procedure for Classification**



- In reality, it is generally not advisable to compare the error of the logistic loss in classification problems
- Rather, we compute the fraction of data which has been misclassified

# Part II: Model Selection and Training/Cross Validation/Test Set

**Choosing a Model** (a BAD method)

- We have 10 models to be placed in consideration, we then find the error of the test set of each model and compare
- Intuitively, we would choose the model with the lowest error
- However, it is probably an overly optimistic estimate of generalization on training data. Therefore, this method is not reccomended

**Dividing the Dataset**

- Now we add another set, called the cross validation to our procedure
- By convention, the ratio should be 60/20/20


We then compute the error for each dataset correspondingly, remember to omit the regularization term

**Model Selection**

- Choose the model with the least error on the cross validation set
- Report the estimate generalization error using the test set

**Model Selection** (neural network)

- The same procedure occurs, we choose the model with the lowest cross validation error
- In this case, we can compute the fraction of misclassified data, and report the error using the test set
