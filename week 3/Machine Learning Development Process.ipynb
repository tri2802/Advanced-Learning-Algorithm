# Part IV: Transfer Learning: Using Data From a Different Task

**Transfer Learning**


- Option 1: only train the output layers parameters
- Option 2: train all parameters
- We can take a pre-trained neural network and fine-tuned it to meet our own goals

**Why does transfer learning work**


- A pre-trained neural network already learns how to recognize the basics of, in this case, any computer vision task
- We can add layers, or modify the output layer so that it can serves our own purpose, i.e.: numerical classification, image processing, etc

**Transfer Learning Summary**
1) Download neural network parameters pretrained on a large dataset with the same input type (audio, speech, image, etc) as our application
2) Further train (fine-tune) the network on our own data
