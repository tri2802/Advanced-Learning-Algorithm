# Part I: Advanced Optimization

**Gradient Descent**


- In the first case, gradient descent will take each small step towards convergence, then we can increase alpha for this algorithm to operate faster
- However, we cannot increase alpha in the second case, and it is moved back and forth; therefore we need to decrease alpha to take smaller steps

**Adam Algorithm Intuition**  

- Adam: Adaptive Moment estimation (not just one $\alpha$)
- For each parameter, Adam will choose automatically the suitable learning rate


- If $w_{j}$ or b keeps moving in the same direction, increase $a_{j}$
- If they keep osciliating, decrease $a_{j}$

**MNIST Adam**
- During compilation, specify the parameter optimizer.  
- model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e_3), loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True))

# Part II: Additional Layer Type

**Dense Layer**

- Each neuron output is a function of all the activation outputs of the previous layer  
- $a^{[k+1]}_{j} = \vec{w_{j}^{[k+1]}} \cdot \vec{a^{[k]} + b_{j}^{[k+1]}}$

**Convolutional Layer**

Instead of each neuron being a pixel in this 3x3 image, we can assign them to look at only a section of the input

- Each neuron only looks at part of the previos layer's output
- Reasons:  
Faster computations, less neurons  
Need less training data (less prone to overfitting)

**Convolutional Neural Network**: the network in which two or more convolutional layer is used


- EKG: heart rate
- let y be the height of the rate, and x be time

- Instead of using 100 inputs of time as being the input layer, we can choose each neuron to look at a subset of time
- The second layer looks at certain subset of the first layer, but not all
- The last layer is a typical classification problem
