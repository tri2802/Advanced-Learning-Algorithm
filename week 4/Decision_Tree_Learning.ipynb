#Part I: Measuring Purity

**Entropy as a measure of impurity**
image.png

- When there is half cats and dogs, impurity will be the highest
- On the contrary, when there is almost all cats or dogs, it will be more pure

- Let $p_{0}$ be the entropy of non-cats, then we have: $p_{0} = 1 - p_{1}$
- Entropy is defined as:
$$H(p_{1}) = -p_{1} log_{2}(p_{1}) - p_{0} log_{2}(p_{0})$$
- By convention, all logs are of based 2
- Note: 0 log 0 = 0

###Part III: Putting it together

**Decision Tree Learning**
- Start with all examples at the root node
- Calculate information gain for all features and pick the highest one
- Split dataset according to selected feature, create left and right branches of the tree
- Keep repeating until one of the criteria is met:  
When a node is 100% one class  
When splitting a node resulting in the tree exceeding max depth  
Information gain from splits is below the threshold  
When number of examples in a node is below the threshold
